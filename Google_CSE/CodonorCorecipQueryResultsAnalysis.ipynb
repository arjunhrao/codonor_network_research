```python
%matplotlib inline

import matplotlib.pyplot
import pylab
import pandas as pd
import numpy as np

dfRes = pd.read_csv('./co_donor_all_query_results_data.csv')
dfResCorecip = pd.read_csv('./co_recip_all_query_results_data.csv')

x = dfRes['num_codonors']
y0 = dfRes['numCommon']
y1 = dfRes['jacInd']
y2 = dfRes['var1']
y3 = dfRes['var2']

#print(y1.count(0))

#y = dfResCorecips['num_corecips']
#x = [1,2,3,4]
#y = [3,4,8,6]
matplotlib.pyplot.scatter(x,y1)
matplotlib.pyplot.show()

matplotlib.pyplot.scatter(x,y2)
matplotlib.pyplot.show()

matplotlib.pyplot.scatter(x,y3)
matplotlib.pyplot.show()


matplotlib.pyplot.scatter(y1,x)
matplotlib.pyplot.show()

matplotlib.pyplot.scatter(y2,x)
matplotlib.pyplot.show()

matplotlib.pyplot.scatter(y3,x)
matplotlib.pyplot.show()
```

```python
from scipy import stats

print(np.corrcoef(x, y0) )
print(np.corrcoef(x, y1) )
print(np.corrcoef(x, y2) )
print(np.corrcoef(x, y3) )

print(stats.describe(x) )
print(stats.describe(y0))
print(stats.describe(y1))
```
